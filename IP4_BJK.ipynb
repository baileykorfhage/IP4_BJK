## Question 5

```python
cols_to_check=['Make','Colour','Odometer (KM)','Doors','Price']
for col in cols_to_check:
    car[col] = car[col].map(lambda x: str(x).replace('$',''))
```

```python
for col in cols_to_check:
    car[col] = car[col].map(lambda x: str(x).replace(',',''))
```

```python
car.head()
```

```python
car['Price']=car['Price'].astype('int64')
```

## Question 6

```python
in_arr= np.array(car['Price'])
out_sum = np.cumsum(in_arr,dtype='int64') 
car.insert(5, "Total Sales", out_sum, True)
```

```python
car.head()
```

## Question 12

```python
fig=plt.figure(figsize=(12,12))

ax1=fig.add_subplot(221)
ax1.set_title('Price vs. total sales line plot')

ax1.plot(car['Price'],car['Total Sales'])
ax1.set_xlabel('Price')
ax1.set_ylabel('Total Sales')

ax2=fig.add_subplot(222)
ax2.set_title('Odometer (KM) vs. Price scatter plot')

ax2.scatter(car['Odometer (KM)'],car['Price'])
ax2.set_xlabel('Odometer (KM)')
ax2.set_ylabel('Price')

ax3=fig.add_subplot(223)
ax3.set_title('Make vs. Odometer (KM) bar plot')

ax3.bar(car['Make'],car['Odometer (KM)'])
ax3.set_xlabel('Make')
ax3.set_ylabel('Odometer (KM)')

ax4=fig.add_subplot(224)
ax4.set_title('Histogram of car make')

ax4.hist(car['Make'])
ax4.set_xlabel('Make')

plt.show()
```
